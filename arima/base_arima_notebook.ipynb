```python
# import numpy as np
from pandas_datareader import data as pdr
import datetime
import matplotlib.pyplot as plt
from base_arima import ARIMAForecaster
```

```python
# 1) Define time window
start = datetime.datetime(2000, 1, 1)
end = datetime.datetime(2024, 12, 31)

# 2) Download target and exogenous series from FRED
#    - TOTDODNS: Total domestic deposits at commercial banks (months)
#    - M2SL: M2 Money Stock (months)
deposit = pdr.DataReader("TOTDODNS", "fred", start, end)
m2 = pdr.DataReader("M2SL", "fred", start, end)

# 3) Prepare series: align indices and names
deposit = (
    deposit.rename(columns={"TOTDODNS": "deposits"}).asfreq("M").fillna(method="ffill")
)
m2 = m2.rename(columns={"M2SL": "m2"}).asfreq("M").fillna(method="ffill")

# 4) Split into train and future exog
#    Use last 24 months of m2 as future exog for demonstration
exog = m2.loc[deposit.index]
exog_future = exog[-24:].copy()

# 5) Instantiate and fit model (trying both with and without Box-Jenkins)
forecaster = ARIMAForecaster(deposit["deposits"], exog=exog)
res = forecaster.fit(p=1, d=1, q=1, P=None, D=None, Q=None, m=12, box_jenkins=False)

# 6) Report the fitted results
forecaster.report()

# 7) Forecast next 24 months
y_pred, ci = forecaster.predict(24, exog_future=exog_future)

# 8) Plot forecast
plt.figure(figsize=(10, 4))
plt.plot(deposit.index, deposit["deposits"], label="Historical Deposits")
plt.plot(y_pred.index, y_pred, "--", label="Forecast Deposits")
if ci is not None:
    plt.fill_between(ci.index, ci.iloc[:, 0], ci.iloc[:, 1], alpha=0.3)
plt.title("Forecast of Bank Deposits with M2 as Exogenous")
plt.legend()
plt.show()

# 9) Diagnostics
forecaster.diagnostics()
```
